#### Handling outliers

```python
#plotting boxplot to check the outliers
df.plot(kind='box',subplots=True,layout=(2,4),figsize=(10,10),title='Outlier Visualization')
plt.show()
```

Outliers found in two columns Compactness and Asymmetry.Coeff 

```python
plt.boxplot(df["Compactness"])
plt.show()
```

```python
#outlier handling using iqr method
Q1=df["Compactness"].quantile(0.25)
Q2=df["Compactness"].quantile(0.50)
Q3=df["Compactness"].quantile(0.75)
IQR=Q3-Q1
low_lim=Q1-1.5*IQR
up_lim=Q3+1.5*IQR
print(low_lim)          
print(up_lim)
```

```python

outlier=[]
for x in df['Compactness']:
    if((x>up_lim)or(x<low_lim)):
        outlier.append(x)
        
len(outlier)
```

```python
#Removing the outliers
ind1=df['Compactness']<low_lim
df.loc[ind1].index
df.drop(df.loc[ind1].index,inplace=True)
plt.boxplot(df['Compactness'])
plt.title('Box plot of Compactness')
```

outlier handling of column Asymmetry.Coeff

```python
plt.boxplot(df["Asymmetry.Coeff"])
plt.show()
```

```python
Q1=df["Asymmetry.Coeff"].quantile(0.25)
Q2=df["Asymmetry.Coeff"].quantile(0.50)
Q3=df["Asymmetry.Coeff"].quantile(0.75)
IQR=Q3-Q1
low_lim=Q1-1.5*IQR
up_lim=Q3+1.5*IQR
print(low_lim)          
print(up_lim)
```

```python

outlier=[]
for x in df['Asymmetry.Coeff']:
    if((x>up_lim)or(x<low_lim)):
        outlier.append(x)
        
len(outlier)
```

```python
#Removing the outliers
ind1=df['Asymmetry.Coeff']>up_lim
df.loc[ind1].index
df.drop(df.loc[ind1].index,inplace=True)
plt.boxplot(df['Asymmetry.Coeff'])
plt.title('Box plot of Asymmetry.Coeff')
plt.show()
```

* Outliers of both features are dropped

#### Train-Test Splitting

```python
# the dataset is splitted into test and train data
#splitting the dataset into X and y
y=df['Type']
X=df.drop('Type',axis=1)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.3)

```

* 30% of the data is test data and 70% of data is train data

# INSIGHTS

* This dataset contains seven features for three different types of wheat kernels
* This particular dataset has 199 entries
* No missing values found
* Outliers found in two columns Compactness and Asymmetry.Coeff
* Outliers of Compactness and Asymmetry.Coeff are dropped
* Noise is found in the column Kernel.Groove with the help of scatterplot,so the index of the noise is found and it is dropped.
* Even though the features are correlated ,feature reduction is not done since the number of total features  is less.
* Type 3 has low kernel.Length,Kernel.Width,Area,Perimeter.
* Type 2 has high kernel.Length,Kernel.Width,Area,Perimeter.
* Data is split into 30% test data and 70% train data.

# MODELLING

### logistic regression

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(fit_intercept=True)
model.fit(X_train,y_train)
y_test_pred=model.predict(X_test)
y_test_prob=model.predict_proba(X_test)
```

```python
model
```

```python
predict=model.predict(X_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score, classification_report
```

```python
# CLASSIFICATION REPORT
print(classification_report(y_train,predict))
```

```python
#Comfusion metrix
cm_rbf=confusion_matrix(y_test,y_test_pred)
print(cm_rbf)
```

```python
# calculatthe accuracy score of the model
accuracy = accuracy_score(y_test,y_test_pred)
print("Accuracy:", accuracy)
```

Accuracy is found to be 98%

```python
# Making a prediction
new_input = [[16.63,15.46,0.87,6.05,3.46,2.04,5.87]]
output = model.predict(new_input)
print(output)
```
